# IPL Exploratory Data Analysis
Seasons 2008–2024

```python
import sys, os
sys.path.insert(0, os.path.join('..', ''))
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from src.data_loader import load_matches, load_deliveries

sns.set_theme(style='darkgrid')
matches = load_matches()
deliveries = load_deliveries()
print(f'Matches: {len(matches):,} | Deliveries: {len(deliveries):,}')
```

## 1. Matches Overview

```python
matches.head()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
season_counts = matches.groupby('season').size()
season_counts.plot(kind='bar', ax=axes[0], color='steelblue', title='Matches per Season')

winner_counts = matches['winner'].value_counts().head(8)
winner_counts.plot(kind='barh', ax=axes[1], color='coral', title='Most Match Wins')
plt.tight_layout()
plt.show()
```

## 2. Toss Analysis

```python
matches['toss_win_match_win'] = matches['toss_winner'] == matches['winner']
toss_effect = matches.groupby('toss_decision')['toss_win_match_win'].mean() * 100
toss_effect.plot(kind='bar', color=['#e94560', '#0f3460'], figsize=(8, 4),
                 title='Win % after winning toss (by decision)', rot=0, ylabel='Win %')
plt.tight_layout()
plt.show()
```

## 3. Top Batsmen

```python
from src.analysis import top_batsmen, top_bowlers
bat = top_batsmen(deliveries)
fig = px.bar(bat, x='batsman', y='runs', color='strike_rate',
             color_continuous_scale='Blues', title='Top 15 Run Scorers',
             labels={'batsman': 'Batsman', 'runs': 'Runs', 'strike_rate': 'SR'})
fig.show()
```

## 4. Top Bowlers

```python
bowl = top_bowlers(deliveries)
fig = px.scatter(bowl, x='wickets', y='economy', size='overs',
                 text='bowler', color='wickets',
                 color_continuous_scale='Reds',
                 title='Wickets vs Economy Rate',
                 labels={'wickets': 'Wickets', 'economy': 'Economy'})
fig.update_traces(textposition='top center')
fig.show()
```

## 5. Run Rate by Over

```python
over_avg = deliveries.groupby('over')['total_runs'].mean()
over_avg.plot(kind='bar', figsize=(14, 4), color='teal',
              title='Average Runs per Ball by Over', ylabel='Avg Runs')
plt.tight_layout()
plt.show()
```
